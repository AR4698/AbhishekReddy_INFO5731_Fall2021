```python
#  Number of Words
import pandas as pd
import math

# **the dataset you collected from assignment two**

train = pd.read_csv('file_name.csv') 
train['word_count'] = train['Review Text'].apply(lambda x: len(str(x).split(" ")))
train[['Review Text','word_count']].head()
```

```python
train['Review Text']
```

```python
train['Review Text']=train['Review Text'].fillna("") # Replace all NaN elements with 0s.
```

```python
train['Review Text'] = train['Review Text'].apply(lambda x: " ".join(x.lower() for x in x.split()))
train['Review Text'].head()
```

```python
#  Removing Punctuation
train['Review Text'] = train['Review Text'].str.replace('[^\w\s]','')
train['Review Text'].head()
```

```python
# Removal of Stop Words
from nltk.corpus import stopwords
stop = stopwords.words('english')
train['Review Text'] = train['Review Text'].apply(lambda x: " ".join(x for x in x.split() if x not in stop))
train['Review Text'].head()
```

```python
# Common word removal
freq = pd.Series(' '.join(train['Review Text']).split()).value_counts()[:10]
freq
```

```python
freq = list(freq.index)
train['Review Text'] = train['Review Text'].apply(lambda x: " ".join(x for x in x.split() if x not in freq))
train['Review Text'].head()
```

# (1) Count the frequency of all the N-grams (N=3).

```python


import collections
import itertools
from nltk import ngrams
from nltk.tokenize import word_tokenize
```

```python
tokens = []
for s in train['Review Text']:
  tokens.append(word_tokenize(s))
k = [x for x in tokens if x != []]
i = list(itertools.chain.from_iterable(k))
trigram = nltk.trigrams(i)
frequency_trigram = nltk.FreqDist(trigram)
frequency_trigram
```

# (2) Calculate the probabilities for all the bigrams in the dataset by using the fomular count(w2 w1) / count(w2). For example, count(really like) / count(really) = 1 / 3 = 0.33.


```python
from collections import Counter
bigram = nltk.bigrams(i)
frequency_trigram = nltk.FreqDist(bigram)
bigram_dict = dict(frequency_trigram)
for word in bigram_dict:
  print( str(word) + ':' + str(bigram_dict[word] / i.count(word[0])))
```

# **Question 2: Undersand TF-IDF and Document representation**

(40 points). Starting from the documents (all the reviews, or abstracts, or tweets) collected for assignment two, write a python program: 

(1) To build the **documents-terms weights (tf*idf) matrix bold text**.

(2) To rank the documents with respect to query (design a query by yourself, for example, "An Outstanding movie with a haunting performance and best character development") by using **cosine similarity**.


```python
train['Review Text']=train['Review Text'].fillna("")
```

```python
import math
def tf__idf(s,w):
  sz = len(s.split(" "))
  tf = s.count(w)/sz
  idf = 0
  if(tf!=0):
    idf = math.log(sz)/s.count(w)
  else:
    return 0;
  return tf*idf

sentences = train['Review Text'].values.tolist()
token_list = set([j for i in sentences for j in i.split(" ")])
tfidf = pd.DataFrame(token_list,columns=["token"])
c = 0
for i in sentences:
  tfidf[str(c)] = tfidf["token"].apply(lambda x : tf__idf(i,x))
  c+=1

tfidf
```
