```python
review_containers = soup.find_all('div', class_ = 'review-container')
```

```python
print(len(review_containers))
```

```python
User_Name=[]
Star=[]
Review_title=[]
Review_Text=[]
Review_Posted_Time=[]
```

```python
soup.find('div',class_='content').div.text
```

```python
oo=soup.find('div', class_ = 'review-container').find('div', class_ = 'ipl-ratings-bar').span.span.text
print(oo)
```

```python
ooo=soup.find('div', class_ = 'review-container').find('div',class_='content').div.text
print(ooo)
```

```python
oooo=soup.find('div', class_ = 'review-container').find('span',class_='review-date').text
print(oooo)
```

```python
for i in review_containers:
    if i.find('div', class_ = 'ipl-ratings-bar') is not None:
        # For User_Names
        N= i.find('div',class_='display-name-date').a.text
        User_Name.append(N)
        # For Star
        S=i.span.span.text
        Star.append(S)
        # For Review_title
        R_T=i.a.text
        Review_title.append(R_T)
        #Review_Text
        R_Text=i.find('div',class_='content').div.text
        Review_Text.append(R_Text)
        #Review_Posted_Time
        R_P_T=i.find('span',class_='review-date').text
        Review_Posted_Time.append(R_P_T)
        
        
```
