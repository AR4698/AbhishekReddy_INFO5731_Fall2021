```python
from bs4 import BeautifulSoup
import requests
import re
import urllib.request
reviews = "https://www.imdb.com/title/tt4154796/reviews"
page = urllib.request.urlopen(reviews)
```

```python
soup = BeautifulSoup(page,'html.parser')
for links in soup.find_all('div',{'class':'lister-item-content'}):
    link = links.find_all('a').get('href')
    print(link)
```

```python
# Using function “prettify” to look at nested structure of HTML page to get into details.

print(soup.prettify())
```
